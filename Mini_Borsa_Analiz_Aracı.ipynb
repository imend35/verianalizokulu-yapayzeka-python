#### AdÄ±m 1: Veri SimÃ¼lasyonu ve Pandas DataFrame OluÅŸturma

GerÃ§ek bir API'den veri Ã§ekmek yerine, Python sÃ¶zlÃ¼kleri ve listeleri kullanarak hayali borsa verisi oluÅŸturacaÄŸÄ±z ve bunu bir Pandas DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼receÄŸiz.

```python
import pandas as pd

def veri_olustur(hisse_adi):
    # SimÃ¼le edilmiÅŸ tarih ve fiyatlar
```

#### AdÄ±m 2: Analiz Fonksiyonu (Basit Ä°statistik)

Veri setindeki basit istatistikleri hesaplayan bir fonksiyon yazÄ±n.

```python
def analiz_yap(df):
    en_yuksek = df['Kapanis_Fiyati'].max()
    en_dusuk = df['Kapanis_Fiyati'].min()
    ortalama = df['Kapanis_Fiyati'].mean()
    
    print(f"\n--- {df['Hisse'].iloc[0]} Analiz SonuÃ§larÄ± ---")

    print(f"Ortalama Fiyat: {ortalama:.2f}")
    
    # OOP'den hatÄ±rlayÄ±n: Bir sÃ¶zlÃ¼k dÃ¶ndÃ¼rmek daha iyi bir yaklaÅŸÄ±mdÄ±r.
    return {"max": en_yuksek, "min": en_dusuk}
```

````python
import pandas as pd
import matplotlib.pyplot as plt

def veri_olustur(hisse_adi):
    # SimÃ¼le edilmiÅŸ tarih ve fiyatlar
    tarihler = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'])
    fiyatlar = [150.50, 151.20, 149.80, 152.00, 153.10]

    veri = {
        'Tarih': tarihler,
        'Kapanis_Fiyati': fiyatlar,
        'Hisse': hisse_adi
    }

    # DataFrame oluÅŸturma
    df = pd.DataFrame(veri)
    df = df.set_index('Tarih') # Tarih sÃ¼tununu index olarak ayarlarÄ±z
    return df
````

````python
def analiz_yap(df):
    en_yuksek = df['Kapanis_Fiyati'].max()
    en_dusuk = df['Kapanis_Fiyati'].min()
    ortalama = df['Kapanis_Fiyati'].mean()

    print(f"\n--- {df['Hisse'].iloc[0]} Analiz SonuÃ§larÄ± ---")
    print(f"En YÃ¼ksek Fiyat: {en_yuksek:.2f}")
    print(f"En DÃ¼ÅŸÃ¼k Fiyat: {en_dusuk:.2f}")
    print(f"Ortalama Fiyat: {ortalama:.2f}")

    # OOP'den hatÄ±rlayÄ±n: Bir sÃ¶zlÃ¼k dÃ¶ndÃ¼rmek daha iyi bir yaklaÅŸÄ±mdÄ±r.
    return {"max": en_yuksek, "min": en_dusuk}
````

````python
 def grafik_ciz(df):
    plt.figure(figsize=(10, 5))

    # Ã‡izgi GrafiÄŸi
    plt.plot(df.index, df['Kapanis_Fiyati'], marker='o', linestyle='-', color='green')

    plt.title(f"{df['Hisse'].iloc[0]} Hisse Fiyat GrafiÄŸi")
    plt.xlabel("Tarih")
    plt.ylabel("KapanÄ±ÅŸ FiyatÄ± ($)")
    plt.grid(True)

    plt.show()
````

````python
# --- Program BaÅŸlangÄ±cÄ± ---
hisse_sembolu = "AAPL" # Ã–rnek hisse sembolÃ¼

# 1. Veriyi Al/OluÅŸtur
hisse_df = veri_olustur(hisse_sembolu)

# 2. Analizi Yap
sonuclar = analiz_yap(hisse_df)

# 3. GrafiÄŸi Ã‡iz
grafik_ciz(hisse_df)
````
