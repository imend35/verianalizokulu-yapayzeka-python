## Python Temelleri: Hızlı Özet

| Konu Başlığı | Ana Kavram | Nerede Kullanılır? |
| :--- | :--- | :--- |
| **1. Değişkenler & Veri Tipleri** | Veriyi hafızada tutan etiketlerdir. Temel tipler: `int`, `float`, `str`, `bool`. | Tüm programlarda, her türlü bilgiyi saklamak için. |
| **2. Operatörler** | Aritmetik (`+`, `-`, `*`), Karşılaştırma (`==`, `>`) ve Mantıksal (`and`, `or`) işlemler yapar. | Matematiksel hesaplamalar ve karar verme mekanizmalarında. |
| **3. Kontrol Akışı** | Programın hangi yoldan ilerleyeceğini belirler. **`if/elif/else`** ile karar verir. | Şartlara bağlı işlemleri yürütmek için. |
| **4. Döngüler** | Bir kod bloğunu tekrar tekrar çalıştırır. **`for`** (listeyi gezmek), **`while`** (koşul doğru olduğu sürece). | Tekrarlayan görevleri otomatikleştirmek için. |
| **5. Veri Yapıları** | Birden fazla veriyi düzenli tutar. **Liste (`[]`)**, **Sözlük (`{key:value}`)**, Demet ve Küme. | Veri gruplarını yönetmek ve organize etmek için. |
| **6. Fonksiyonlar & Modüller**| **Fonksiyonlar** (`def`) tekrar kullanılabilir kod bloklarıdır. **Modüller** (`import`) ise kod organizasyonudur. | Kod tekrarını azaltmak ve projeyi düzenlemek için. |
| **7. Hata Yönetimi** | Hata oluşabilecek kodları **`try`** içine alır; hatayı **`except`** ile yakalar ve programın çökmesini önler. | Programın dayanıklılığını ve güvenilirliğini artırmak için. |
| **8. Nesne Yönelimli Prog. (OOP)** | Programlamayı gerçek nesneler üzerinden yapar. **Sınıflar** şablon, **Nesneler** somut örnektir. | Büyük ve karmaşık yazılım projelerini yönetmek için. |

## Popüler Python Kütüphaneleri ve Kullanım Alanları

Python'ın asıl gücü, geliştirilmiş hazır modüller (kütüphaneler) ekosisteminden gelir. Neredeyse çözülmüş her problem için bir kütüphane bulabilirsiniz.

### 1. Web Geliştirme

| Kütüphane | Amaç | Kısa Açıklama |
| :--- | :--- | :--- |
| **Django** | Tam kapsamlı (Full-Stack) web çatısı. | Büyük ve ölçeklenebilir web uygulamaları, içerik yönetim sistemleri ve API'ler oluşturmak için kullanılır. Hızlı ve güvenlidir. |
| **Flask** | Hafif ve minimal web çatısı. | Daha küçük projeler, prototipler ve basit API hizmetleri geliştirmek için idealdir. Daha fazla esneklik sunar. |

### 2. Veri Bilimi ve Analizi

| Kütüphane | Amaç | Kısa Açıklama |
| :--- | :--- | :--- |
| **NumPy** | Sayısal hesaplamalar ve hızlı dizi (array) işlemleri. | Özellikle çok boyutlu dizilerle çalışırken temel oluşturur. Tüm bilimsel kütüphanelerin temelidir. |
| **Pandas** | Veri manipülasyonu ve analizi. | Tablo şeklindeki verileri (CSV, Excel vb.) okumak, temizlemek, filtrelemek ve analiz etmek için kullanılır. **Veri Biliminin vazgeçilmezidir.** |
| **Matplotlib / Seaborn** | Veri görselleştirme. | Analiz sonuçlarını grafikler, histogramlar ve çizelgeler aracılığıyla anlaşılır hale getirmek için kullanılır. |

### 3. Yapay Zeka ve Makine Öğrenimi

| Kütüphane | Amaç | Kısa Açıklama |
| :--- | :--- | :--- |
| **Scikit-learn** | Klasik Makine Öğrenimi (ML) algoritmaları. | Regresyon, sınıflandırma, kümeleme gibi temel ML görevlerini kolayca uygulamak için en popüler kütüphanedir. |
| **TensorFlow / PyTorch** | Derin Öğrenme (Deep Learning) çatısı. | Büyük veri kümeleri üzerinde karmaşık sinir ağları (Yapay Zeka) oluşturmak, eğitmek ve dağıtmak için kullanılır. |

### 4. Otomasyon, Test ve Diğer Alanlar

| Kütüphane | Amaç | Kısa Açıklama |
| :--- | :--- | :--- |
| **Requests** | HTTP istekleri yapma. | Web sitelerinden veri çekmek (API kullanımı) ve internet üzerinden iletişim kurmak için kullanılır. |
| **Selenium** | Web tarayıcısı otomasyonu. | Web sitelerini test etmek, otomatik form doldurmak veya belirli görevleri otomatik yapmak için tarayıcıyı programlama. |
| **Beautiful Soup** | Web kazıma (Web Scraping). | HTML/XML dosyalarından veri çıkarmayı kolaylaştırır (Örn: Bir haber sitesinden başlıkları çekmek). |
