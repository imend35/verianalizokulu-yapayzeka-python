## 6\. Fonksiyonlar ve ModÃ¼ller: Kodun DÃ¼zenlenmesi

### 6.1. Fonksiyonlar (Functions)

Fonksiyonlar, belirli bir gÃ¶revi yerine getiren ve isimlendirilmiÅŸ, yeniden kullanÄ±labilir kod bloklarÄ±dÄ±r. Python'daki temel ilke olan **DRY (Don't Repeat Yourself - Kendini Tekrar Etme)** prensibini uygulamamÄ±zÄ± saÄŸlarlar.

#### TanÄ±mlama ve KullanÄ±m

Bir fonksiyon, `def` (define - tanÄ±mla) anahtar kelimesi ile oluÅŸturulur.

  * **Parametre (Parameters):** Fonksiyonun iÃ§ine alabileceÄŸi deÄŸiÅŸkenlerdir.
  * **ArgÃ¼man (Arguments):** Fonksiyon Ã§aÄŸrÄ±lÄ±rken parametrelere verilen deÄŸerlerdir.
  * **`return`:** Fonksiyonun Ã§alÄ±ÅŸmasÄ±nÄ± bitirir ve bir sonucu geri dÃ¶ndÃ¼rÃ¼r.

```python
# Fonksiyon TanÄ±mlama

    sonuc = sayi1 + sayi2
```

Bir deÄŸiÅŸkenin nerede tanÄ±mlandÄ±ÄŸÄ±, onun nerede kullanÄ±labileceÄŸini belirler:

  * **Global (KÃ¼resel) DeÄŸiÅŸkenler:** FonksiyonlarÄ±n dÄ±ÅŸÄ±nda tanÄ±mlanmÄ±ÅŸtÄ±r ve programÄ±n her yerinden eriÅŸilebilir.

### 6.3. ModÃ¼ller (Modules)

ModÃ¼l, basitÃ§e Python kodlarÄ± iÃ§eren bir dosyadÄ±r (`.py` uzantÄ±lÄ±). Bir modÃ¼l, baÅŸka bir programda kullanÄ±labilecek fonksiyonlarÄ±, sÄ±nÄ±flarÄ± veya deÄŸiÅŸkenleri iÃ§erir.

ModÃ¼ller, kodunuzu mantÄ±ksal olarak bÃ¶lmenize ve dÃ¼zenlemenize yardÄ±mcÄ± olur.

#### KullanÄ±m: `import`

Python'da halihazÄ±rda var olan birÃ§ok kullanÄ±ÅŸlÄ± modÃ¼l (kÃ¼tÃ¼phane) bulunur.

```python

# random modÃ¼lÃ¼nden randint fonksiyonunu kullanma

print(f"Rastgele sayÄ±: {rastgele_sayi}")
```

#### `from ... import ...`

ModÃ¼l adÄ±nÄ± tekrar tekrar yazmamak iÃ§in modÃ¼l iÃ§indeki belirli bir fonksiyonu doÄŸrudan iÃ§e aktarabilirsiniz.

```python

```
